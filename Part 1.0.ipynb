```python
def mean(L):
    return(sum(L)/len(L))

def SD(L):
    variance = sum([((x - mean(L)) ** 2) for x in L]) / len(L)
    res = variance ** 0.5
    return (res)
```

```python
xdata_SD=[]
for i in range(len(xdata)):
    L=[]
    L.append((xdata[i][0]-mean(T))/SD(T))
    L.append((xdata[i][1]-mean(I))/SD(I))
    L.append((xdata[i][2]-mean(R))/SD(R))
    xdata_SD.append(L)
```

```python
X = np.array(xdata_SD)#define array
print (X)
print (X.T)  #print the transpose
Xmean = np.mean(X,0)  # compute mean vector
print (Xmean)
```

```python
C = np.cov(X.T)  #transpose is matrix we want to work with - compute covariance matrix
print (C)
```

```python
w, v = LA.eig(C)  # get the eigenvalues w and the eigenvectors 
print (w)
print (v)
```
